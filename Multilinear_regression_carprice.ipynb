DATA PRE-PROCESSING Encoding the categorical feature

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['name']=le.fit_transform(x['name'])
x['fueltypes']=le.fit_transform(x['fueltypes'])
x['aspiration']=le.fit_transform(x['aspiration'])
x['doornumbers']=le.fit_transform(x['doornumbers'])
x['carbody']=le.fit_transform(x['carbody'])
x['drivewheels']=le.fit_transform(x['drivewheels'])
x['enginelocation']=le.fit_transform(x['enginelocation'])
x['enginetype']=le.fit_transform(x['enginetype'])
x['cylindernumber']=le.fit_transform(x['cylindernumber'])
x['fuelsystem']=le.fit_transform(x['fuelsystem'])
```

```python
x
```

standardization/normalization

```python
#StandardScaler method using standadisation
from sklearn.preprocessing import StandardScaler
standardisation=StandardScaler()
res=standardisation.fit_transform(x)
res
```

```python
from sklearn import preprocessing
x1=preprocessing.scale(x)
x1
```

```python
y
```

TRAIN/TEST SPLIT
                1.split the daataset in two parts test and train
                2.Train the model in training set
                3.Test the model in testing set

```python
# train,test X,Y split model selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(res,y,test_size=0.3,random_state=42)
```

```python
y_train
```

IMPORT THE MODELS & MODEL TRAINING

```python
# train model ,learn
from sklearn.linear_model import LinearRegression
train_model=LinearRegression().fit(x_train,y_train)
train_model
```

PREDICTION ON TEST DATA

```python
y_prediction=train_model.predict(x_test)
y_prediction
```

```python
y_test
```

Evaluation Metrics For a regression problem, we have to find the following metrics

1.R squared
       2.MSE
       3.RMSE
       4.MAE

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
R2 = r2_score(y_test,y_prediction)
MAE = mean_absolute_error(y_test,y_prediction)
MSE = mean_squared_error(y_test,y_prediction)
RMSE = np.sqrt(MSE)
```

```python
print('R squared is',R2,'\n','MAE is',MAE,'\n','RMSE is',RMSE,'\n','MSE is',MSE)
```
